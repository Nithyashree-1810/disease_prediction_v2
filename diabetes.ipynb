```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix,roc_curve,auc
import pickle

# Step 1: Load the dataset
# Replace 'path_to_your_dataset.csv' with the actual path to your dataset
diabetes_db = pd.read_csv('/content/diabetes (1).csv')

# Step 2: Inspect the dataset
print("First 5 rows of the dataset:")
print(diabetes_db.head())

print("\nShape of dataset:", diabetes_db.shape)
print("\nSummary statistics:")
print(diabetes_db.describe())

print("\nClass distribution in 'Outcome':")
print(diabetes_db['Outcome'].value_counts())

# Step 3: Split the dataset into features (X) and target (y)
x = diabetes_db.drop(columns='Outcome', axis=1)  # Drop the Outcome column
y = diabetes_db['Outcome']  # Outcome is the target

# Step 4: Standardize the feature data
scaler = StandardScaler()
scaler.fit(x)  # Fit the scaler on the feature data
x = scaler.transform(x)  # Transform the feature data

# Step 5: Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, stratify=y, random_state=2
)

print(f"\nTraining data shape: {x_train.shape}, Testing data shape: {x_test.shape}")

# Step 6: Train an SVM classifier
classifier = svm.SVC(kernel='linear')  # Linear kernel for SVM
classifier.fit(x_train, y_train)

# Step 7: Evaluate the model
# Training data accuracy
x_train_prediction = classifier.predict(x_train)
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
training_data_precision = precision_score(y_train, x_train_prediction)
training_data_recall = recall_score(y_train, x_train_prediction)
training_data_f1 = f1_score(y_train, x_train_prediction)
print(f"\nTraining data accuracy: {training_data_accuracy:.2f}")
print(f"Training data precision: {training_data_precision:.2f}")
print(f"Training data recall: {training_data_recall:.2f}")
print(f"Training data F1-score: {training_data_f1:.2f}")
# Testing data accuracy
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(y_test, x_test_prediction)
test_data_precision = precision_score(y_test, x_test_prediction)
test_data_recall = recall_score(y_test, x_test_prediction)
test_data_f1 = f1_score(y_test, x_test_prediction)
print(f"Testing data accuracy: {test_data_accuracy:.2f}")
print(f"Testing data precision: {test_data_precision:.2f}")
print(f"Testing data recall: {test_data_recall:.2f}")
print(f"Testing data F1-score: {test_data_f1:.2f}")

# Step 8: Save the trained model
filename = 'diabetes_model.sav'
pickle.dump(classifier, open(filename, 'wb'))
print(f"\nModel saved to '{filename}'")

# Step 9: Make predictions based on input values
# Example input values for prediction
input_data = (5, 166, 72, 19, 175, 25.8, 0.587, 51)  # Replace with your values
input_data_as_numpy_array = np.asarray(input_data).reshape(1, -1)
print("Processed input data:", input_data)

 # Reshape for prediction
std_data = scaler.transform(input_data_as_numpy_array)  # Standardize input

# Prediction using the trained model
prediction = classifier.predict(std_data)

# Display prediction result
if prediction[0] == 0:
    print("\nPrediction: The person is NOT diabetic (Outcome: 0)")
else:
    print("\nPrediction: The person IS diabetic (Outcome: 1)")



# Step 10: Load the saved model and test it
loaded_model = pickle.load(open(filename, 'rb'))
loaded_model_prediction = loaded_model.predict(std_data)
if loaded_model_prediction[0] == 0:
    print("\nLoaded model prediction: The person is NOT diabetic (Outcome: 0)")
else:
    print("\nLoaded model prediction: The person IS diabetic (Outcome: 1)")

# Optional: Predict multiple rows
# Example batch prediction
batch_input = [
    (5, 166, 72, 19, 175, 25.8, 0.587, 51),  # Input 1
    (2, 110, 70, 15, 85, 24.5, 0.500, 32),  # Input 2
]
batch_input = np.asarray(batch_input)
batch_std_data = scaler.transform(batch_input)
batch_predictions = classifier.predict(batch_std_data)

print("\nBatch predictions:")
for i, prediction in enumerate(batch_predictions):
    result = "NOT diabetic (Outcome: 0)" if prediction == 0 else "IS diabetic (Outcome: 1)"
    print(f"Input {i + 1}: {result}")

filename = 'diabetes_model.sav'
pickle.dump(classifier, open(filename, 'wb'))
print(f"\nModel saved to '{filename}'")
```
